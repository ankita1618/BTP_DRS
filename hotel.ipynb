```python
def get_loc(address, position, maps_key='KEY'):
    maps_api_url = 'https://maps.googleapis.com/maps/api/geocode/json'
    request_url = maps_api_url + '?' + urllib.parse.urlencode({'address':address,'key':maps_key})
    response = requests.get(request_url)
    resp_json_payload = response.json()
    print(response.json())
    if(len(resp_json_payload['results'])==0):
        return 'NaN'
    out = resp_json_payload['results'][0]['geometry']['location']
    if position == 'latitude':
        return float(out['lat'])
    elif position == 'longitude':
        return float(out['lng'])
```

```python
latitude =[]
longitude = []
for i in hotel['Name']:
    address = i+',Leh'
    latitude.append(get_loc(address,'latitude'))
    longitude.append(get_loc(address,'longitude'))
    
```

```python
hotel['latitude'] = latitude
hotel['longitude'] = longitude
```
